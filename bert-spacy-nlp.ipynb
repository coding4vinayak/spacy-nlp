```python
import spacy
from spacy.training import Example

# Load the transformer model
nlp = spacy.load("en_core_web_trf")

# Example text for processing
text = "Apple is looking at buying U.K. startup for $1 billion"

# Process the text
doc = nlp(text)

# Named Entity Recognition
print("Entities in the text:")
for ent in doc.ents:
    print(ent.text, ent.label_)

# Part-of-Speech Tagging
print("\nTokens and POS tags:")
for token in doc:
    print(token.text, token.pos_)

# Fine-tuning for custom tasks (e.g., text classification)
# Add a text classifier
if "textcat" not in nlp.pipe_names:
    textcat = nlp.add_pipe("textcat", last=True)
else:
    textcat = nlp.get_pipe("textcat")

textcat.add_label("POSITIVE")
textcat.add_label("NEGATIVE")

# Prepare training data
train_data = [
    ("I love this movie!", {"cats": {"POSITIVE": 1.0, "NEGATIVE": 0.0}}),
    ("I hate this movie.", {"cats": {"POSITIVE": 0.0, "NEGATIVE": 1.0}})
]

# Start the training
optimizer = nlp.begin_training()

# Train for a number of iterations
for epoch in range(10):  # Adjust number of iterations as needed
    losses = {}
    for text, annotations in train_data:
        doc = nlp.make_doc(text)
        example = Example.from_dict(doc, annotations)
        nlp.update([example], losses=losses, drop=0.5)
    print(f"Epoch {epoch} Losses: {losses}")

# Evaluate the model
texts = ["I really enjoyed this book.", "This product is terrible.", "i hate maya", "i love rishi"]
docs = [nlp(text) for text in texts]

for doc in docs:
    print(f"Text: {doc.text}")
    print(f"Prediction: {doc.cats}")

```

```python
from transformers import BertTokenizer, BertForSequenceClassification, Trainer, TrainingArguments
from transformers import DataCollatorWithPadding
from datasets import load_dataset, Dataset, DatasetDict
import torch

# Prepare the dataset
train_texts = ["I love this movie!", "I hate this movie.", "This book is amazing!", "I don't like this book."]
train_labels = [1, 0, 1, 0]
val_texts = ["I really enjoyed this book.", "This product is terrible.", "I hate Maya.", "I love Rishi."]
val_labels = [1, 0, 0, 1]

train_data = {'text': train_texts, 'label': train_labels}
val_data = {'text': val_texts, 'label': val_labels}

# Convert to Hugging Face datasets
train_dataset = Dataset.from_dict(train_data)
val_dataset = Dataset.from_dict(val_data)

# Load the tokenizer and model
tokenizer = BertTokenizer.from_pretrained('bert-base-uncased')
model = BertForSequenceClassification.from_pretrained('bert-base-uncased', num_labels=2)

# Tokenize the dataset
def tokenize_function(examples):
    return tokenizer(examples['text'], truncation=True, padding=True)

train_dataset = train_dataset.map(tokenize_function, batched=True)
val_dataset = val_dataset.map(tokenize_function, batched=True)

# Set format for PyTorch tensors
train_dataset.set_format(type='torch', columns=['input_ids', 'attention_mask', 'label'])
val_dataset.set_format(type='torch', columns=['input_ids', 'attention_mask', 'label'])

# Define training arguments
training_args = TrainingArguments(
    output_dir='./results',
    evaluation_strategy="epoch",
    learning_rate=2e-5,
    per_device_train_batch_size=8,
    per_device_eval_batch_size=8,
    num_train_epochs=3,
    weight_decay=0.01,
)

# Data collator
data_collator = DataCollatorWithPadding(tokenizer=tokenizer)

# Initialize Trainer
trainer = Trainer(
    model=model,
    args=training_args,
    train_dataset=train_dataset,
    eval_dataset=val_dataset,
    tokenizer=tokenizer,
    data_collator=data_collator,
)

# Train the model
trainer.train()

# Evaluate the model
eval_results = trainer.evaluate()
print(f"Evaluation results: {eval_results}")

# Save the model
model.save_pretrained('./text_classification_model')
tokenizer.save_pretrained('./text_classification_model')

# Load the saved model and tokenizer for inference
loaded_model = BertForSequenceClassification.from_pretrained('./text_classification_model')
loaded_tokenizer = BertTokenizer.from_pretrained('./text_classification_model')

# Inference
def predict(text):
    inputs = loaded_tokenizer(text, return_tensors='pt', truncation=True, padding=True)
    outputs = loaded_model(**inputs)
    probs = torch.nn.functional.softmax(outputs.logits, dim=-1)
    return probs

test_texts = ["I really enjoyed this book.", "This product is terrible.", "I hate Maya.", "I love Rishi."]
for text in test_texts:
    probs = predict(text)
    print(f"Text: {text}")
    print(f"Prediction: {probs}")
```

```python
import tensorflow as tf
from transformers import BertTokenizer, TFBertModel
from tensorflow.keras.layers import Input, Flatten, Dense
from tensorflow.keras.models import Model

# Example data
texts = ["I love this movie!", "I hate this movie.", "This book is amazing!", "I don't like this book."]
labels = [1, 0, 1, 0]  # Assuming 1 for positive sentiment and 0 for negative sentiment

# Load BERT tokenizer and model
tokenizer = BertTokenizer.from_pretrained('bert-base-uncased')
bert_model = TFBertModel.from_pretrained('bert-base-uncased')

# Tokenize input texts
tokenized_texts = tokenizer(texts, padding=True, truncation=True, return_tensors="tf")

# Extract BERT embeddings
bert_outputs = bert_model(tokenized_texts)
last_hidden_states = bert_outputs.last_hidden_state  # Shape: (batch_size, sequence_length, hidden_size)

# Define sequential model
input_layer = Input(shape=(last_hidden_states.shape[1], last_hidden_states.shape[2]))
flatten_layer = Flatten()(input_layer)
dense_layer = Dense(64, activation='relu')(flatten_layer)
output_layer = Dense(1, activation='sigmoid')(dense_layer)

model = Model(inputs=input_layer, outputs=output_layer)

# Compile the model
model.compile(optimizer='adam', loss='binary_crossentropy', metrics=['accuracy'])

# Train the model
model.fit(last_hidden_states, labels, epochs=5, batch_size=2, verbose=1)

# Evaluate the model (optional)
# evaluation_loss, evaluation_accuracy = model.evaluate(validation_data)
# print("Evaluation Loss: ", evaluation_loss)
# print("Evaluation Accuracy: ", evaluation_accuracy)

# Inference (predict on new data)
new_texts = ["I really enjoyed this book.", "This product is terrible."]
tokenized_new_texts = tokenizer(new_texts, padding=True, truncation=True, return_tensors="tf")
new_bert_outputs = bert_model(tokenized_new_texts)
new_last_hidden_states = new_bert_outputs.last_hidden_state
predictions = model.predict(new_last_hidden_states)
print("Predictions: ", predictions)
```
