```python
import spacy
from collections import Counter
from string import punctuation

# Load spaCy's English language model
nlp = spacy.load("en_core_web_sm")

def summarize_text(text, num_sentences=3):
    # Process the text using spaCy
    doc = nlp(text)

    # Extract sentences and words
    sentences = [sent for sent in doc.sents]
    words = [token.text.lower() for token in doc if token.text.lower() not in nlp.Defaults.stop_words and token.text not in punctuation]

    # Get word frequencies
    word_freq = Counter(words)

    # Calculate sentence scores based on word frequencies
    sentence_scores = {}
    for sent in sentences:
        for word in sent:
            if word.text.lower() in word_freq:
                if sent not in sentence_scores:
                    sentence_scores[sent] = word_freq[word.text.lower()]
                else:
                    sentence_scores[sent] += word_freq[word.text.lower()]

    # Get the most important sentences
    important_sentences = sorted(sentence_scores, key=sentence_scores.get, reverse=True)[:num_sentences]

    # Combine the sentences to form the summary
    summary = " ".join([sent.text for sent in important_sentences])
    
    return summary

# Example usage
text = """
The quick brown fox jumps over the lazy dog. The quick brown fox is known for its agility and speed. 
It can leap over large obstacles with ease. Foxes are generally found in the wild, but they can also 
adapt to urban environments. The quick brown fox has become a popular character in various cultural references. 
Lazy dogs, on the other hand, are often seen lounging around and taking it easy. They are not known for 
their speed or agility. Despite their differences, the quick brown fox and the lazy dog can sometimes be 
seen interacting with each other in stories and folklore.
"""

summary = summarize_text(text, num_sentences=1)
print("Summary:", summary)
```
